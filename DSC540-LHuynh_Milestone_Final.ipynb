### Cleaning/Formatting Website Data
Due to the lack of success with the data source suggested in Milestone 1, I switched to using this Wikipedia HTML instead: https://en.wikipedia.org/wiki/Immigration_to_the_United_States

```python
# Import libraries:
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Fetch and parse the web page
url = "https://en.wikipedia.org/wiki/Immigration_to_the_United_States"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all tables
tables = soup.find_all("table", {"class": "wikitable"})

# Print the type of data_table 
type(tables)
```

```python
# List all tables 
for i, table in enumerate(tables):
    # Print the table index (starting from 1) and a preview of the first 5 rows
    print(f"Table {i + 1}")
    print(pd.read_html(str(table))[0].head())
    print("\n")
```

After fetching the Wikipedia page, parsing it with BeautifulSoup for "wikitable" class tables, and printing both the type and the list of tables, I selected two tables for analysis:
- Table 4: Persons obtaining legal permanent resident status by fiscal year
- Table 7: Inflow of New Legal Permanent Residents by region, 2015–2022

The next step is to convert these tables into pandas DataFrames and perform transformations on them.

```python
# Transform 'table 4' into pandas dataframe df1
df1 = pd.read_html(str(tables[3]))[0]
print("Persons obtaining legal permanent resident status by fiscal year:")
df1
```

```python
# Step 1: Extract and rename columns
years = []
values = []

# Identify columns related to years and values
for col in df1.columns:
    if 'Year' in col:
        years.append(col) # 'years' list collects columns containing 'Year'
    elif 'Unnamed' in col:  
        values.append(col) # 'values' list collects columns that are 'Unnamed'

# Step 2: Combine into a new DataFrame
# Create an empty DataFrame 'df1_new'
df1_new = pd.DataFrame()

# Loop through pairs of 'years' and 'values' columns
for year_col, value_col in zip(years, values):
    # For each pair, create a temporary DataFrame 'df1_temp' with columns 'Year' and 'Person (Count)'
    df1_temp = pd.DataFrame({
        'Year': df1[year_col],
        'Person (Count)': df1[value_col]
    })
    # Concatenate 'df1_temp' to 'df1_new'
    df1_new = pd.concat([df1_new, df1_temp])

# Step 3: Sort 'df1_new' by Year column
df1_new.sort_values(by='Year', inplace=True)

# Step 4: Reset index
df1_new.reset_index(drop=True, inplace=True)

# Step 5: Convert 'Year' column to numeric type, coercing any errors
df1_new['Year'] = pd.to_numeric(df1_new['Year'], errors='coerce')

# Display the transformed DataFrame
print("Transformed DataFrame:")
df1_new.head()
```

```python
# Step 6: Format numbers with commas and handle decimals
def format_value(x):
    # Check if 'x' is an instance of int or float
    if isinstance(x, (int, float)):
        # Determine the formatting pattern based on the type and value of 'x'
        # If x is an integer (isinstance(x, int)) or a float that represents a whole number (x.is_integer()), 
        # format x without any decimal places using '{:,.0f}'.
        # If x is a float with decimals, format x with two decimal places using '{:,.2f}'.
        formatted_value = '{:,.0f}' if isinstance(x, int) or x.is_integer() else '{:,.2f}'
        # Format 'x' using the determined pattern and remove trailing zeros and decimal point 
        return formatted_value.format(x).rstrip('0').rstrip('.')
    else:
        # If 'x' is not an int or float, return it as is
        return x


# Apply formatting to 'Person (Count)' column
df1_new['Person (Count)'] = df1_new['Person (Count)'].apply(format_value)

# Display transformed DataFrame
print("Transformed DataFrame:")
print("Persons obtaining legal permanent resident status by fiscal year")
df1_new.tail()
```

```python
# Transform 'table 7' into pandas dataframe df2
df2 = pd.read_html(str(tables[6]))[0]
print("Inflow of New Legal Permanent Residents by region, 2015–2022:")
df2
```

```python
# Step 1: Reshape the data and initial cleaning
# Transpose and reset index
df2_new = df2.T.reset_index() 
# Set first row as column headers
df2_new.columns = df2_new.iloc[0]
# Skip first row and reset index
df2_new = df2_new[1:].reset_index(drop=True)
# Rename 'Region' column to 'Year'
df2_new = df2_new.rename(columns={'Region': 'Year'}) 

# Step 2: Replace cells in 'Year' column that contain '% of total' with NaN
df2_new['Year'] = df2_new['Year'].replace(r'% of total', pd.NA, regex=True)

# Step 3: Remove rows where 'Year' is NaN
df2_new = df2_new.dropna(subset=['Year'])

# Step 4: Handle Special Characters in 'Year' column
df2_new['Year'] = df2_new['Year'].str.replace(r'\[.*\]', '', regex=True)

# Step 5: Convert data in the remaining columns to numeric and exclude 'Year' column, 
# which contains categorical data 
numeric_columns = df2_new.columns.drop('Year')

# Loop through each numeric column
for col in numeric_columns:
    # Convert each column to numeric, coercing errors to NaN
    df2_new[col] = pd.to_numeric(df2_new[col], errors='coerce')
    # Apply formatting: add commas and remove decimals
    df2_new[col] = df2_new[col].apply(lambda x: "{:,}".format(int(x)) if pd.notnull(x) else x)

# Step 6: Remove rows with any NaN values
df2_new = df2_new.dropna()

# Step 7: Reset index to be sequential
df2_new = df2_new.reset_index(drop=True)

# Display the transformed DataFrame
print("Transformed DataFrame:")
print("Inflow of New Legal Permanent Residents by region, 2015–2022:")
df2_new
```

Two tables from Wikipedia's Immigration to the United States page were utilized for analysis. For the table on "Persons obtaining legal permanent resident status by fiscal year" data transformation included extracting and renaming columns, combining data into a new DataFrame, sorting by year, converting the year column to numeric type while handling errors, and formatting numbers for clarity. In the case of "Inflow of New Legal Permanent Residents by region, 2015–2022" steps involved reshaping the data, handling special characters in the year column, converting numeric data while excluding categorical columns, removing rows with missing values, and resetting the index to ensure sequential order. These transformations were essential for preparing the data for further analysis and interpretation.

Source: https://data.census.gov/table/ACSST1Y2017.S0502?q=Native%20and%20Foreign-Born%20in%20state%202017&t=Class%20of%20Worker

```python
import json
import requests
import pandas as pd

# Path to the JSON file containing the API key
json_file_path = '/Users/linhhuynh/Desktop/DSC540/acs_key.json'

# Open the JSON file and load the API key
with open(json_file_path) as f:
    keys = json.load(f)
    acs_key = keys['acs_key']

# Base URL for the Census API for the given table
base_url = 'https://api.census.gov/data/2017/acs/acs1/subject'

# Parameters for the API request
params = {
    'get': 'GEO_ID,NAME,group(S0502)',  
    'for': 'state:31',  # Filter data for Nebraska (state code 31)
    'key': acs_key
}

# Send a GET request to the API
response = requests.get(base_url, params=params)

# Check if the request was successful
if response.status_code == 200:
    print("API request successful.")
    data = response.json()
    
    # Convert JSON data to a Pandas DataFrame
    df = pd.DataFrame(data[1:], columns=data[0])
    
    # Save the raw DataFrame to a CSV file
    df.to_csv('nebraska_acs_data_raw.csv', index=False)
    print("Raw data saved to 'nebraska_acs_data_raw.csv'.")
else:
    print(f"API request failed with status code {response.status_code}")
    print(response.text) 
```

```python
# Visualization #4: Immigrant Employment by Occupation in Nebraska (2017)
# Calculate the number of people in each occupation category
merged_df2['Professional_num'] = merged_df2['TotalPop'] * (merged_df2['Professional'] / 100)
merged_df2['Service_x_num'] = merged_df2['TotalPop'] * (merged_df2['Service_x'] / 100)
merged_df2['Office_num'] = merged_df2['TotalPop'] * (merged_df2['Office'] / 100)
merged_df2['Construction_num'] = merged_df2['TotalPop'] * (merged_df2['Construction'] / 100)
merged_df2['Production_num'] = merged_df2['TotalPop'] * (merged_df2['Production'] / 100)

# Plot the bar chart
occupations = ['Professional', 'Service_x', 'Office', 'Construction', 'Production']
employment = [sum(merged_df2['Professional_num']), sum(merged_df2['Service_x_num']), sum(merged_df2['Office_num']), 
              sum(merged_df2['Construction_num']), sum(merged_df2['Production_num'])]

plt.figure(figsize=(10, 6))
plt.bar(occupations, employment, color='pink')
plt.xlabel('Occupation')
plt.ylabel('Number of People')
plt.title('Immigrant Employment by Occupation in Nebraska (2017)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization #5: Income Distribution
# Choose income columns
income = ['Income1To9999','Income10kTo14999','Income15kTo24999','Income25kTo34999',
                    'Income35kTo49999','Income50kTo74999','Income75kPlus']

# Clean and convert columns
for column in income:
    merged_df2[column] = merged_df2[column].astype(str)  # Ensure column is treated as string
    merged_df2[column] = merged_df2[column].str.replace('%', '', regex=False)  # Remove '%'
    merged_df2[column] = pd.to_numeric(merged_df2[column], errors='coerce')  # Convert to numeric
```

```python
# Calculate mean for each column
means = merged_df2[income].mean()

# Plot the pie chart
plt.figure(figsize=(10, 8))
plt.pie(means, labels=means.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Income Distribution')
plt.show()
```
